### melon 차트 만들기

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
{"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36"}
```

```python
m_url = "https://www.melon.com/chart/index.htm"
h = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36"}
res = req.get(m_url, headers = h)
soup = bs(res.text, 'lxml')
song = soup.select('.rank01 > span > a')
singer = soup.select('.rank02 > a:first-child')
# :nth-child(숫자)  가져오고 싶은 번째를 지정할 수 있다.
```

```python
for i in song:
    print(i.text)
```

```python
song_list = []
singer_list = []
rank_list = []

for i in range(len(song)):
    song_list.append(song[i].text)
    singer_list.append(singer[i].text)
    rank_list.append(i+1)
```

```python
song_list
```

```python
melon_dic = {'곡명' : song_list, '가수' : singer_list, '순위' : rank_list}
melon = pd.DataFrame(melon_dic)
```

```python
melon.set_index('순위', inplace = True)
```

```python
melon
```

```python
# 출력되는 dataframe 의 길이 늘리기
pd.set_option('display.max_row', 100)
```

```python
melon.to_csv('./melon.csv', encoding = 'euc-kr')

#  ./  -> 현재 폴더에 저장   ../  -> 상위 폴더에 저장
```

### 여러 페이지 리뷰 가져오기

```python
title_list = []
review_list = []

for i in range(1,5):

    url = 'https://movie.naver.com/movie/point/af/list.nhn?&page='+str(i)
    res = req.get(url)
    soup = bs(res.text, 'lxml')

    title = soup.select('.color_b')
    review = soup.select('.title')

    a = soup.select('a')
    div = soup.select('.list_netizen_score')

    for i in a:
        i.extract()

    for i in div:
        i.extract()
    
    for i in range(len(title)):
        title_list.append(title[i].text.strip())
        review_list.append(review[i].text.strip())
```

```python
review_dic = {'제목' : title_list , '리뷰' : review_list,}
review = pd.DataFrame(review_dic)
```

### tqdm_notebook >> 진행사항을 볼 수 있는 함수

```python
from tqdm import tqdm_notebook
```

```python
title_list = []
score_list = []
rank_list = []
date_list = []

for day in tqdm_notebook(days):

    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&tg=0&date='+day
    res = req.get(url)
    soup = bs(res.text, 'lxml')

    title = soup.select('.tit5')
    score = soup.select('.point')

    for i in range(len(title)):
        title_list.append(title[i].text.strip())
        score_list.append(score[i].text.strip())
        rank_list.append(i+1)
        date_list.append(day)
```

```python
score_dic = {'순위' : rank_list , '제목' : title_list , '평점' : score_list, '날짜' : date_list}
score = pd.DataFrame(score_dic)
```

```python
score.set_index('순위', inplace = True)
score
```
