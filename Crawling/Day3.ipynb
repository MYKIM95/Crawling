### 2015년부터 2020년 데이터 가져오기

```python
result_list = []
date_list = []
rank_list = []

for day in range(2015,2021):
    url = "https://sports.news.naver.com/kbaseball/record/index.nhn?category=kbo&year=" + str(day)
    res = req.get(url)
    soup = bs(res.text, 'lxml')

    # 가져올 개수 제한 걸어주기
    result = soup.select('td', limit = 110)
    
    
    for j in range(len(result)):
        result_list.append(result[j].text.strip())

    for k in range(1, 11):
        date_list.append(day)
        rank_list.append(k)
```

```python
arr = np.array(result_list)
farr = arr.reshape(-1,11)
farr
```

```python
col = ['팀', '경기수', '승', '패', '무', '승률', '게임차', '연속', '출루율', '장타율', '최근 10경기']
df = pd.DataFrame(farr, columns = col)
df
```

```python
dic = {'순위' : rank_list, '날짜' : date_list}

df2 = pd.DataFrame(dic)

fdf = pd.concat([df,df2], axis = 1)

fdf.set_index('순위', inplace = True)

fdf
```
